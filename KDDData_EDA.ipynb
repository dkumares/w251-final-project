```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
colNames = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'land', 'wrong_fragment', 'urgent',
             'hot', 'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell', 'su_attempted', 'num_root',
             'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login', 'is_guest_login', 'count',
             'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate',
             'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate', 'dst_host_diff_srv_rate',
             'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate',
             'dst_host_rerror_rate', 'dst_host_srv_rerror_rate', 'Label']
df = pd.read_csv("./Data/kddcup.data.csv", names= colNames)

# To display the top 5 rows
df.head(5)
```

```python
# Plotting a Histogram
df.service.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("network service on the destination, e.g., http, telnet, etc.")
plt.ylabel('Number of services')
plt.xlabel('service type');
```

```python
# Plotting a Histogram
df.Label.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Labels predicted on the network data")
plt.ylabel('Number of labels')
plt.xlabel('Label');
```

```python
# Count the number of labels across every feature
df.groupby('Label').count()
```

```python
df['Label'].value_counts()
```

```python
# Finding the relations between the variables.
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Plotting a scatter plot between a couple of features that are correlated according to the heatmap above
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['num_root'], df['num_compromised'])
ax.set_xlabel('number of root accesses')
ax.set_ylabel('number of compromised conditions')
plt.show()
```
